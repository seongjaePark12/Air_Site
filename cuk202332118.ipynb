```python
from IPython import get_ipython
from IPython.display import display
# %%
!pip install python-Levenshtein # Install the Levenshtein library

import pandas as pd
import Levenshtein
from google.colab import files

# 파일 업로드
uploaded = files.upload()
file_path = "/content/ChatbotData.csv"  # 업로드된 파일 경로

# 학습 데이터 CSV 파일 로드
def load_data(file_path):
    """ CSV 파일을 로드하고 DataFrame으로 반환하는 함수 """
    return pd.read_csv(file_path)

# 레벤슈타인 거리 계산 함수
def levenshtein_distance(str1, str2):
    """ 두 문자열 간의 레벤슈타인 거리를 계산하는 함수 """
    return Levenshtein.distance(str1, str2)

# 가장 유사한 질문의 인덱스 찾기
def find_most_similar_question(chat_question, dataset):
    """ Chat 질문과 가장 유사한 학습 데이터 질문의 인덱스를 찾는 함수 """
    # Changed "question" to "Q" to match the actual column name in the CSV
    similarities = [levenshtein_distance(chat_question, q) for q in dataset["Q"]]
    most_similar_index = similarities.index(min(similarities))
    return most_similar_index

# 답변 출력 함수
def get_answer(chat_question, dataset):
    """ 가장 유사한 질문의 인덱스를 찾아 해당 답변을 반환하는 함수 """
    index = find_most_similar_question(chat_question, dataset)
    # Changed "answer" to "A" to match the actual column name in the CSV
    return dataset.loc[index, "A"]

# 실행 예시
if __name__ == "__main__":
    dataset = load_data(file_path)

    chat_question = input("질문을 입력하세요: ")  # 사용자 입력
    answer = get_answer(chat_question, dataset)

    print(f"가장 유사한 질문의 인덱스: {find_most_similar_question(chat_question, dataset)}")
    print(f"답변: {answer}")
```
